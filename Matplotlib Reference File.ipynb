```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Line Plot

```python
x = np.arange(0,10,0.1)
y = 2*x + 5

plt.plot(x,y)
plt.show()
```

## Customizing Line Plot

```python
x = np.arange(0,10,1)
y = 2*x + 5
y1 = 5*x + 6

plt.figure(figsize=(10,5))
plt.plot(x, y, linewidth=2.0, linestyle="--", color="#0f8f4f", alpha=0.5, marker="x")
plt.plot(x, y1, linewidth=2.0, linestyle=":", color="orange", alpha=0.8, marker="o")
plt.title("Title for plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend(["Line 1", "Line 2"], loc="best")
plt.grid(b=True, which='major', axis='both')
plt.show()
```

### Ploting more than one graph in one figure

```python
x1 = np.arange(0,10,1)
y = 2*x + 5
y1 = 3*x + 10
y2 = 4*x + 15
y3 = 5*x + 20

# Create two or more plots in a figure
plt.figure(figsize=(25,20))
plt.subplot(2,2,1)
plt.plot(x1, y, linestyle="--", linewidth=1.8, color='red', alpha=.5, marker="o")
plt.xlabel("X axis 1")
plt.ylabel("Y axis 1")
plt.title("Graph 1")

plt.subplot(2,2,2)
plt.plot(x1, y1, linestyle=":", linewidth=1.8, color='yellow', alpha=.8, marker="+")
plt.xlabel("X axis 2")
plt.ylabel("Y axis 2")
plt.title("Graph 2")

plt.subplot(2,2,3)
plt.plot(x1, y1, linestyle="--", linewidth=1.8, color='green', alpha=.5, marker="x")
plt.xlabel("X axis 3")
plt.ylabel("Y axis 3")
plt.title("Graph 3")

plt.subplot(2,2,4)
plt.plot(x1, y1, linestyle="--", linewidth=1.8, color='black', alpha=.5, marker="o")
plt.xlabel("X axis 4")
plt.ylabel("Y axis 4")
plt.title("Graph 4")
plt.show()
```

## How to create a Bar Chart

```python
# Creating dataset
data = {"Apples":20, "Mangoes":15, "Lemon":30, "Orange":10}
fruits = list(data.keys())
quantity = list(data.values())

plt.figure(figsize=(15,10))
plt.bar(fruits, quantity)
plt.title("Quantity of fruits left in stock")
plt.xlabel("Fruits")
plt.ylabel("Quantity")
plt.show()
```

## Customizing Bar Chart

```python
# Creating dataset
data = {"Apples":20, "Mangoes":15, "Lemon":30, "Orange":10}
fruits = list(data.keys())
quantity = list(data.values())

plt.figure(figsize=(15,10))

# Creating Bar plot
plt.barh(fruits, quantity, color="orange")
plt.title("Quantity of fruits left in stock")
plt.xlabel("Quantity")
plt.ylabel("Fruits")
plt.grid(True)
plt.show()
```

## How to create a Scatter plot

```python
x = np.arange(10,81,10, dtype=np.int32)
a = np.random.randint(1,6,8)
b = np.random.randint(1,6,8)

# Create scatter plot
plt.scatter(x,a)
plt.scatter(x,b)
plt.grid(True)
plt.show()
```

## Customizing a Scatter plot

```python
x = np.arange(10,81,10, dtype=np.int32)
a = np.random.randint(1,6,8)
b = np.random.randint(1,6,8)

# Create scatter plot
plt.scatter(x, a, c="green", s=150, edgecolor='black', marker="1", alpha=1)
plt.scatter(x, b, c="red", s=150, edgecolor='black', marker="4", alpha=1)
plt.legend(["a","b"], loc="best")
plt.title("Scatter Plot Demo")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid(True)
plt.savefig("Scatter Demo.png")
plt.show()
```

## How to plot a Histogram
#### The total number of values found within the boundaries of each bin (range) is plotted against each bin.

```python
numbers = np.random.randint(10,100,24)
bins = np.arange(0,101,20)

plt.hist(numbers, bins=bins)
plt.grid(True)
plt.show()
print(sorted(numbers))
```

## Customizing Histograms

```python
numbers = np.random.randint(10,100,24)
bins = np.arange(0,101,20)

# Creating the histogram
plt.hist(numbers, bins=bins, edgecolor="#000000", color="blue")
plt.title("Histogram Plot Demo")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid(True)
plt.savefig("Histogram Plot Demo.png")
plt.show()
```

## How to create a Box (Whisker) Plot. 

```python
total = np.random.randint(1,101,10)
order = np.random.randint(1,101,10)
discount = np.random.randint(1,101,10)

data = list([total, order, discount])

plt.boxplot(data, showmeans=True)
plt.title("Box Plot Demo")
plt.grid(True)
plt.show()
```

## How to create a Violin Plot
### It's shaded area is used to find the probabilty density function for the dataset

```python
total = np.random.randint(1,101,10)
order = np.random.randint(1,101,10)
discount = np.random.randint(1,101,10)

data = list([total, order, discount])

# ploting the data
plt.figure(figsize=(15,10))
ax1 = plt.subplot()
ax1.violinplot(data, showmeans=True, showmedians=True)
plt.title("Violin Plot Demo")
plt.grid(True)

plt.show()
```

## How to create a Pie Chart

```python
label = ["Dogs", "Cats", "Wolf", "Lion"]
size  = np.random.randint(10,100,4)

plt.figure(figsize=(15,10))
plt.pie(size, labels=label, autopct="%1.1f%%", shadow=True, startangle=9, explode=(0.05,0.05,0.05,0.05))
plt.title("Pie Chart Demo")
plt.show()
```

## How to create a Doughnut Chart

```python
group_names = ["Group1","Group2","Group3"]
group_sizes = [20,40,60]
size_centre = [5]

a,b,c = [plt.cm.Blues, plt.cm.Greens, plt.cm.Reds]

# Creating Doughnut Chart
pie1 = plt.pie(group_sizes, labels=group_names, radius=1.5, colors=[a(0.5),b(1.5),c(1.0)])
pie2 = plt.pie(size_centre, radius=1.0, colors="w")
plt.show()
```

## How to create Exploded Doughnut Chart

```python
label = ["Dogs", "Cats", "Wolf", "Lion"]
size  = np.random.randint(10,100,4)

size_centre = [1]

plt.figure(figsize=(15,10))
plt.pie(size, labels=label, autopct="%1.1f%%", radius=1, shadow=True, startangle=9, explode=(0.05,0.05,0.05,0.05))
plt.pie(size_centre, radius=.4, colors="w")
plt.title("Pie Chart Demo")
plt.show()
```

## Area Plot

```python
x = np.arange(0,11,1)
y = 2*x + 5

plt.stackplot(x, y, colors="teal", alpha=.3)
plt.plot(x, y, linewidth=1, linestyle="-", color="blue", alpha=1, marker="o")
plt.title("Area Plot Demo")
plt.show()
```
